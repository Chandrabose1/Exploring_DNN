```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

```python
from sklearn.datasets import make_circles
```

```python
n_samples=1000
x,y=make_circles(n_samples=1000,noise=0.01,random_state=42)
```

```python
print(x)
```

```python
circle_x=x[:,0]
circle_y=x[:,1]
```

```python
x_min,x_max=circle_x.min(),circle_x.max()
y_min,y_max=circle_y.min(),circle_y.max()
```

```python
x_cord,y_cord=np.meshgrid(np.linspace(x_min,x_max+0.1,100),np.linspace(y_min,y_max+0.1,100))
```

```python
x_in=np.c_[x_cord.ravel(),y_cord.ravel()]
```

```python
model=tf.keras.Sequential([tf.keras.layers.Dense(10),
                          tf.keras.layers.Dense(1)])
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
             optimizer=tf.keras.optimizers.SGD(),
             metrics=["accuracy"])
model.fit(x,y,epochs=100,verbose=1)
```

```python
ypred=model.predict(x_in)
```

```python
ypred = np.round(ypred).reshape(x_cord.shape)
```

```python
ypred = np.round(ypred).reshape(x_cord.shape)
```

```python
plt.contourf(x_cord,y_cord,ypred)
plt.scatter(circle_x,circle_y,c=y, s=40, cmap=plt.cm.RdYlBu)
```

```python
ypred[-5:-1,:]
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection='3d')
#ax.contour3D(x_cord,y_cord,ypred,zdir='x')
ax.contour3D(x_cord,y_cord,ypred,zdir='y')
#ax.contour3D(x_cord,y_cord,ypred,zdir='z')
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('c')
plt.show()
```
